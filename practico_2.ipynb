```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
MAT_RGB2YIQ = np.array([[0.299, 0.587, 0.114],
                        [0.596,-0.275,-0.321],
                        [0.211,-0.523, 0.311]])

def apply_matrix(img, M):
    return np.matmul(img.reshape((-1,3)), M.T).reshape(img.shape)

def rgb2yiq(img):
    return apply_matrix(img, MAT_RGB2YIQ)

def yiq2rgb(img):
    return apply_matrix(img, np.linalg.inv(MAT_RGB2YIQ))
```

```python
img_1 = imageio.imread('imageio:coffee.png')[:, 50:550, :]/255
img_2 = imageio.imread('imageio:astronaut.png')[56:456, 6:506, :]/255

fig, axes = plt.subplots(1, 2, figsize=(10,5))
axes[0].imshow(img_1)
axes[1].imshow(img_2)
```

## 1. Aritmética de pixels

Implementar las siguientes operaciones aritméticas:

1. Suma y diferencia clampeada en RGB (la suma ya está hecha en el primer ejemplo)
2. Suma y diferencia promediada en RGB
3. Suma y diferencia en YIQ
4. If-lighter / If-darker

```python
# Implementación de suma clampleada
img_sum_clamp = np.clip(img_1 + img_2, 0, 1)

fig, axes = plt.subplots(1, 3, figsize=(15,5))
axes[0].imshow(img_1)
axes[1].imshow(img_2)
axes[2].imshow(img_sum_clamp)
axes[2].set_title('Suma clampeada')

# Implementación de diferencia clampeada 
img_dif_clamp = np.clip(img_1 - img_2, 0, 1)
fig, axes = plt.subplots(1, 3, figsize=(15,5))
axes[0].imshow(img_1)
axes[1].imshow(img_2)
axes[2].imshow(img_dif_clamp)
axes[2].set_title('Diferencia clampeada')
```

```python
# Implementación de suma promediada
img_sum_prom = np.clip((img_1 + img_2)/2, 0, 1)

fig, axes = plt.subplots(1, 3, figsize=(15,5))
axes[0].imshow(img_1)
axes[1].imshow(img_2)
axes[2].imshow(img_sum_prom)
axes[2].set_title('Suma promediada')

# Implementación de diferencia promediada 
img_dif_prom = np.clip((img_1 - img_2)/2, 0, 1)
fig, axes = plt.subplots(1, 3, figsize=(15,5))
axes[0].imshow(img_1)
axes[1].imshow(img_2)
axes[2].imshow(img_dif_prom)
axes[2].set_title('Diferencia promediada')
```

```python
img_1_yiq = rgb2yiq(img_1)
img_2_yiq = rgb2yiq(img_2)

Y1 = img_1_yiq[:,:,0]
I1 = img_1_yiq[:,:,1]
Q1 = img_1_yiq[:,:,2]

Y2 = img_2_yiq[:,:,0]
I2 = img_2_yiq[:,:,1]
Q2 = img_2_yiq[:,:,2]

##
## Implementación de suma en YIQ
##

#clampeada
img_yiq_sumclamp = np.zeros(img_1.shape)

Y3 = Y1 + Y2
I3 = (Y1*I1 + Y2*I2) / (Y1+Y2)
Q3 = (Y1*Q1 + Y2*Q2) / (Y1+Y2)

img_yiq_sumclamp[:,:,0] = Y3
img_yiq_sumclamp[:,:,1] = I3
img_yiq_sumclamp[:,:,2] = Q3

#promediada
img_yiq_sumprom = np.zeros(img_1.shape)

Y3 = (Y1 + Y2) / 2
I3 = (Y1*I1 + Y2*I2) / (Y1+Y2)
Q3 = (Y1*Q1 + Y2*Q2) / (Y1+Y2)

img_yiq_sumprom[:,:,0] = Y3
img_yiq_sumprom[:,:,1] = I3
img_yiq_sumprom[:,:,2] = Q3

##
## Implementación de diferencia en YIQ
##

#clampeada
img_yiq_difclamp = np.zeros(img_1.shape)

Y3 = Y1 - Y2
I3 = (Y1*I1 - Y2*I2) / (Y1+Y2)
Q3 = (Y1*Q1 - Y2*Q2) / (Y1+Y2)

img_yiq_difclamp[:,:,0] = Y3
img_yiq_difclamp[:,:,1] = I3
img_yiq_difclamp[:,:,2] = Q3

#promediada
img_yiq_difprom = np.zeros(img_1.shape)

Y3 = (Y1 - Y2) / 2
I3 = (Y1*I1 - Y2*I2) / (Y1+Y2)
Q3 = (Y1*Q1 - Y2*Q2) / (Y1+Y2)

img_yiq_difprom[:,:,0] = Y3
img_yiq_difprom[:,:,1] = I3
img_yiq_difprom[:,:,2] = Q3


# plots
fig, axes = plt.subplots(1, 3, figsize=(10,5))
axes[0].imshow(img_1)
axes[1].imshow(img_2)
axes[2].imshow(np.clip(yiq2rgb(img_yiq_sumclamp), 0, 1))
axes[2].set_title('Suma clampeada en YIQ')

fig, axes = plt.subplots(1, 3, figsize=(10,5))
axes[0].imshow(img_1)
axes[1].imshow(img_2)
axes[2].imshow(np.clip(yiq2rgb(img_yiq_sumprom), 0, 1))
axes[2].set_title('Suma promediada en YIQ')

fig, axes = plt.subplots(1, 3, figsize=(10,5))
axes[0].imshow(img_1)
axes[1].imshow(img_2)
axes[2].imshow(np.clip(yiq2rgb(img_yiq_difclamp), 0, 1))
axes[2].set_title('Diferencia clampeada en YIQ')

fig, axes = plt.subplots(1, 3, figsize=(10,5))
axes[0].imshow(img_1)
axes[1].imshow(img_2)
axes[2].imshow(np.clip(yiq2rgb(img_yiq_difprom), 0, 1))
axes[2].set_title('Diferencia promediadada en YIQ')

```
