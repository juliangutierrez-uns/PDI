```python
MAT_RGB2YIQ = np.array([[0.299, 0.587, 0.114],
                        [0.596,-0.275,-0.321],
                        [0.211,-0.523, 0.311]])

def apply_matrix(img, M):
    return np.matmul(img.reshape((-1,3)), M.T).reshape(img.shape)

def rgb2yiq(img):
    return apply_matrix(img, MAT_RGB2YIQ)

def yiq2rgb(img):
    return apply_matrix(img, np.linalg.inv(MAT_RGB2YIQ))

def rmse(img1, img2):
    return np.sqrt(np.mean((img1-img2)**2))
```

```python
from matplotlib import cm

def plot_kernel(data, ax=None):
    rows, cols = data.shape
    y, x = np.meshgrid(np.arange(rows),np.arange(cols),indexing='ij')
    if ax == None:
        fig = plt.figure()
        ax = fig.gca(projection='3d')
    _min, _max = (np.min(data), np.max(data))
    ax.plot_surface(x, y, data.T, cmap=cm.jet, vmin=_min, vmax=_max)

def plot_images_and_kernel(img, img_filt, kernel):
    fig = plt.figure(figsize=(17,5))
    ax1 = fig.add_subplot(131)
    ax1.imshow(img, 'gray')
    ax1.title.set_text('Input image')
    ax2 = fig.add_subplot(132)
    ax2.imshow(img_filt, 'gray')
    ax2.title.set_text('Filtered image')
    ax3 = fig.add_subplot(133, projection='3d')
    plot_kernel(kernel, ax=ax3)
    ax3.title.set_text('Kernel')
    plt.show()
```

```python
# Ejemplo del kernel identidad
ker = np.zeros((3,3))
ker[1,1] = 1
ker /= ker.sum()

img_filt = convolve2d(img, ker, 'same')

plot_images_and_kernel(img, img_filt, ker)
```

```python
from scipy.stats import norm

### 1.1 Filtros simétricos

##Pasabajos

# Box
def box(N):
    kernel = np.ones((N,N))
    kernel /= kernel.sum()

    return kernel
    
N = int(input("BOX   .Ingrese tamaño del kernel (NxN): "))
img_filt = convolve2d(img, box(N), 'same')
plot_images_and_kernel(img, img_filt, box(N))

# Circle
def circle(N):
    kernel = np.zeros((N,N))
    
    # circulo
    r = N // 2
    x0 = y0 = N // 2
    y, x = np.ogrid[:N, :N]
    y = y[::-1]
    circle = ((x - x0)**2 + (y-y0)**2)<= r**2
    
    kernel[circle == True] = 1
    kernel /= kernel.sum()
    
    return kernel

N = int(input("CIRCLE   .Ingrese tamaño del kernel (NxN): "))
img_filt = convolve2d(img, circle(N), 'same')
plot_images_and_kernel(img, img_filt, circle(N))

# Barlett
def barlett(N):
    a = np.zeros(N)
    for i in range(N):
        if i <= (N-1)/2:
            a[i] = i+1
        else:
            a[i] = N-i
    
    b = np.reshape(a, (N,1))
    
    kernel = np.zeros((N,N))
    kernel = np.multiply(a,b)
    kernel /= kernel.sum()

    return kernel
    
N = int(input("BARLETT   .Ingrese tamaño del kernel (NxN): "))
img_filt = convolve2d(img, barlett(N), 'same')
plot_images_and_kernel(img, img_filt, barlett(N))

# Gaussian
def gaussian(N,std):
    a = np.zeros(N)
    for i in range(N):
        a[i] = norm.pdf(i,(N-1)/2,std)
        
    b = np.reshape(a,(N,1))
    kernel = np.multiply(a,b)
    kernel /= kernel.sum()
    return kernel
  
N = int(input("GAUSSIAN   .Ingrese tamaño del kernel (NxN): "))
std = float(input("GAUSSIAN   .Ingrese desvío estándar: "))
img_filt = convolve2d(img, gaussian(N,std), 'same')
plot_images_and_kernel(img, img_filt, gaussian(N,std))


## Pasaaltos

# Laplace
def laplace(neighbors):
    if neighbors == 4:
        kernel = np.array([[ 0, -1,  0],
                           [-1,  4, -1],
                           [ 0, -1, 0]])
    elif neighbors == 8:
        kernel = np.array([[-1, -1, -1],
                           [-1,  8, -1],
                           [-1, -1, -1]])
    return kernel

neighbors = int(input("LAPLACE   .Ingrese cantidad de vecinos (4 u 8): "))
img_filt = convolve2d(img, laplace(neighbors), 'same')
img_filt[img_filt<=0] = 0
img_filt[img_filt>=1] = 1
plot_images_and_kernel(img, img_filt, laplace(neighbors))
    
# Pasalto a partir de un pasabajo
def pa_pb(N,std):
    center = int((N-1)/2)
    kernel_id = np.zeros((N,N))
    kernel_id[center,center] = 1
    
    kernel_pb = gaussian(N,std)
    kernel_pa = kernel_id - kernel_pb
    
    return kernel_pa
    
N = int(input("PA-PB (Gaussian)   .Ingrese tamaño del kernel (NxN): "))
std = float(input("PA-PB (Gaussian)   .Ingrese desvío estándar: "))
img_filt = convolve2d(img, pa_pb(N,std), 'same')
img_filt[img_filt<=0] = 0
img_filt[img_filt>=1] = 1
plot_images_and_kernel(img, img_filt, pa_pb(N,std))
    
## Pasabanda

# Difference of Gaussian
def dog(N,std1,std2):
    kernel_g1 = gaussian(N,std1)
    kernel_g2 = gaussian(N,std2)
    
    kernel_dog = kernel_g1 - kernel_g2
    return kernel_dog
    

N = int(input("Difference of Gaussian   .Ingrese tamaño del kernel (NxN): "))
std1 = float(input("Difference of Gaussian   .Ingrese desvío estándar 1: "))
std2 = float(input("Difference of Gaussian   .Ingrese desvío estándar 2: "))
img_filt = convolve2d(img, dog(N,std1,std2), 'same')
img_filt[img_filt<=0] = 0
img_filt[img_filt>=1] = 1
plot_images_and_kernel(img, img_filt, dog(N,std1,std2))    
    
## Otros

# Mejora de contraste
def contraste(neighbors, proporcion):
    kernel_id = np.zeros((3,3))
    kernel_id[1,1] = 1
    
    kernel_l = laplace(neighbors)
    kernel = kernel_id + (np.multiply(proporcion,kernel_l))
    
    return kernel

neighbors = int(input("CONTRASTE   .Ingrese cantidad de vecinos (4 u 8): "))
proporcion = float(input("CONTRASTE   .Ingrese proporción de kernel: "))
img_filt = convolve2d(img, contraste(neighbors,proporcion), 'same')
img_filt[img_filt<=0] = 0
img_filt[img_filt>=1] = 1
plot_images_and_kernel(img, img_filt, contraste(neighbors,proporcion))
```

```python
### 1.2 Filtros direccionales

# Sobel
def sobel(img):
    kernel_gx = np.array([[-1, 0, 1],
                          [-2, 0, 2],
                          [-1, 0, 1]])
    
    kernel_gy = np.array([[ 1,  2,  1],
                          [ 0,  0,  0],
                          [-1, -2, -1]])
    
    gx = convolve2d(img, kernel_gx, 'same')
    gy = convolve2d(img, kernel_gy, 'same')

    return kernel_gx, kernel_gy, gx, gy

kernel_gx, kernel_gy, gx, gy = sobel(img)
abs_gx = abs(gx)
abs_gy = abs(gy)
modulo = np.sqrt((gx**2) + (gy**2))
fase = np.arctan(gy/gx)

# normalizo fase [0,1]
fase[np.isnan(fase)] = np.pi/2 
fase = (fase - fase.min()) / (fase.max() - fase.min())

plot_images_and_kernel(img, abs_gx, kernel_gx)
plot_images_and_kernel(img, abs_gy, kernel_gy)
fig, axes = plt.subplots(1, 2, figsize=(15,5))
axes[0].imshow(modulo, 'gray')
axes[1].imshow(fase, cmap='cividis')
axes[0].set_title('Módulo')
axes[1].set_title('Fase')
plt.colorbar(axes[1].imshow(fase, cmap='cividis'))
```

```python
### 1.3 Detector de bordes a partir de Sobel

# Detector de bordes - umbral
kernel_gx, kernel_gy, gx, gy = sobel(img)
modulo = np.sqrt((gx**2) + (gy**2))
fase = np.arctan(gy/gx)

# normalizo fase [0,1]
fase[np.isnan(fase)] = np.pi/2 
fase = (fase - fase.min()) / (fase.max() - fase.min())

# umbral
umbral = float(input('Umbral: '))
modulo[modulo<umbral] = 0
modulo[modulo>=umbral] = 1

fig, axes = plt.subplots(1, 2, figsize=(15,5))
axes[0].imshow(modulo, 'gray')
axes[1].imshow(fase, cmap='cividis')
axes[0].set_title('Módulo')
axes[1].set_title('Fase')
plt.colorbar(axes[1].imshow(fase, cmap='cividis'))
```

### 1.4 Detector de bordes de canny

El detector de bordes de Canny consta de varias etapas:

 1. Reducción de ruido:
  
      La detección de bordes es susceptible al ruido en la imagen, por lo tanto el primer paso es eliminar el ruido en la imagen con un filtro gaussiano de 5x5.
      
      
 2. Gradiente de imagen:
 
     Luego, a la imagen suavizada (filtro Guassiano) se le aplica un filtro Sobel para obtener los gradientes tanto horizontales Gx, como verticales Gy (transición de luminancias). A partir de estas dos imágenes, se puede determinar la pendiente (módulo) y la dirección (fase) del borde.
     
     
 3. Umbral de magnitud de gradiente o supresión de corte de límite inferior:
 
     Es una técnica de adelgazamiento de bordes, donde se aplica la supresión de corte de límite inferior para encontrar las ubicaciones con el cambio más brusco del valor de intensidad. Se compara la fuerza del borde del píxel actual con la fuerza del borde del píxel en las direcciones de gradiente positivo y negativo. Luego, si la fuerza del borde del píxel actual es la más grande en comparación con los otros píxeles en la máscara con la misma dirección, el valor se se mantiene. De lo contrario, el valor se elimina.
     
     
 4. Umbral de histéresis:
 
     Esta etapa decide cuáles son realmente bordes y cuáles no. Para esto, se necesitan dos valores de umbral, minU y maxU. Cualquier borde con un gradiente de intensidad superior a maxU seguramente será borde ("borde fuerte") y aquellos por debajo de minU seguramente no serán bordes, por lo tanto, se eliminan. Aquellos que se encuentran entre estos dos umbrales se clasifican como bordes o no bordes en función de su conectividad ("bordes débiles"). Si están conectados a píxeles de "borde fuerte", se consideran parte de los bordes. De lo contrario, se eliminan. Esta etapa también elimina los ruidos de los píxeles pequeños asumiendo que los bordes son líneas largas.
